```python
#converting poster column to actual date 

def estimate_posted_date(posted_str):
    today = datetime.today()
    posted_str = str(posted_str).lower()
    if "today" in posted_str or "just now" in posted_str:
        return today
    if "yesterday" in posted_str:
        return today - timedelta(days=1)
    match = re.search(r"(\d+)\s+(day|week|month|hour)", posted_str)
    if match:
        num = int(match.group(1))
        unit = match.group(2)
        if unit == "hour":
            return today
        elif unit == "day":
            return today - timedelta(days=num)
        elif unit == "week":
            return today - timedelta(weeks=num)
        elif unit == "month":
            return today - timedelta(days=num * 30)
    return None
```

```python
#creating simplified columns
df["PostedDate"] = df["Posted"].map(estimate_posted_date)
df["PostedDate"] = pd.to_datetime(df["PostedDate"])
df["Posted_Days_Ago"] = (datetime.today() - df["PostedDate"]).dt.days
```

```python
#cleaning the Experience column to min and max experience columns
def extract_experience(exp_str):
    match = re.findall(r"(\d+)", str(exp_str))
    if len(match) == 2:
        return int(match[0]), int(match[1])
    elif len(match) == 1:
        return int(match[0]), int(match[0])
    return None, None
```

```python
df["MinExperience"], df["MaxExperience"] = zip(*df["Experience"].map(extract_experience))
```

```python
# Check for true nulls (empty, NaN, etc.)
null_counts = df.isnull().sum()
print(null_counts)
```

```python
# Check for textual "not specified", "n/a", "none", "nan" (case-insensitive)
not_specified_counts = df.apply(
    lambda col: col.astype(str).str.lower().str.contains("not specified|n/a|none|nan", na=False)
).sum()
print(not_specified_counts)
```

```python
rows_with_nulls = df[df.isnull().any(axis=1)]
print(rows_with_nulls)
```

```python


df["PostedDate"] = df["PostedDate"].fillna(pd.Timestamp.today())  
df["Posted_Days_Ago"] = df["Posted_Days_Ago"].fillna(0).astype(int)  
df["MinExperience"] = df["MinExperience"].fillna(0).astype(int)
df["MaxExperience"] = df["MaxExperience"].fillna(0).astype(int)

df.fillna("Not Specified", inplace=True)
```

```python
#replacing the not specified skills by the skills above it
df["Skills"] = df["Skills"].replace("Not Specified", pd.NA).ffill()
```

```python
# Only one known placeholder: "Not Specified"
placeholder = "Not specified"

# Replace individual fields based on that value
df.loc[df["Title"] == placeholder, "Title"] = "Data Analyst"
df.loc[df["Location"] == placeholder, "Location"] = "India"

# Use company name as fallback URL
df.loc[df["URL"] == placeholder, "URL"] = df["Company"].apply(
    lambda x: f"https://www.naukri.com/{str(x).strip().replace(' ', '-').lower()}"
)

# Set experience to "0-0 Yrs"
df.loc[df["Experience"] == placeholder, "Experience"] = "0-0 Yrs"
```

```python
#rephrasing experience
df["MinExperience"], df["MaxExperience"] = zip(*df["Experience"].map(extract_experience))
```
